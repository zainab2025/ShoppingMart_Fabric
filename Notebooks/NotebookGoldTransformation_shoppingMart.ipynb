### read Data from silver Layer

```python
from pyspark.sql.functions import *
df_orders = spark.read.parquet("Files/ShoppingMart_Silver_orders/ShoppingMart_customersOrders.csv")
df_orders.write.mode('overwrite').parquet('Files/ShoppingMart_Gold_orders/ShoppingMart_customersOrders.csv')
```

```python
df_web_logs = spark.read.parquet("Files/ShoppingMart_Silver_web_logs/ShoppingMart_web_logs")
df_web_logs = df_web_logs.groupBy('user_id','action','page').count()
df_web_logs.write.mode('overwrite').parquet('Files/ShoppingMart_Gold_web_logs/ShoppingMart_web_logs')

```

```python
df_social_media = spark.read.parquet("Files/ShoppingMart_Silver_social_media/ShoppingMart_social_media")
df_social_media = df_social_media.groupBy('platform','sentiment').count()
df_social_media.write.mode('overwrite').parquet('Files/ShoppingMart_Gold_social_media/ShoppingMart_social_media')
```

```python
df_reviews = spark.read.parquet("Files/ShoppingMart_Silver_Reviews/ShoppingMart_review")
df_reviews = df_reviews.groupby('product_id').agg(avg('rating').alias('average'))
df_reviews.write.mode('overwrite').parquet('Files/ShoppingMart_Gold_Reviews/ShoppingMart_review')
```
